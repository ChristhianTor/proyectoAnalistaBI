```python
#Realizamos la importacion de las librerias
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Conectar a la base de datos
conn = sqlite3.connect('bca_ops.db')
```

```python
#Se realiza el query con el cual se trae la informacion de la DB para los FTD mensuales
query_ftd = """
WITH first_deposits AS (
    SELECT player_id, MIN(deposit_date) AS first_deposit_date
    FROM deposits
    GROUP BY player_id
)
SELECT 
    SUBSTR(fd.first_deposit_date, 1, 7) AS month,
    COUNT(*) AS ftd_count
FROM first_deposits fd
GROUP BY month
ORDER BY month;
"""

#la libreria se encarga de realizar la conexion con la DB y hace la lectura de la consulta
ftd_df = pd.read_sql_query(query_ftd, conn)
```

```python
#Se realiza el query con el cual se trae la informacion de la DB para los CPA mensuales
query_cpa = """
WITH total_deposits AS (
    SELECT player_id, SUM(amount) AS total_amount
    FROM deposits
    GROUP BY player_id
),
qualified_players AS (
    SELECT player_id
    FROM total_deposits
    WHERE total_amount > 100
),
first_deposits AS (
    SELECT player_id, MIN(deposit_date) AS first_deposit_date
    FROM deposits
    GROUP BY player_id
)
SELECT 
    SUBSTR(fd.first_deposit_date, 1, 7) AS month,
    COUNT(*) AS cpa_count
FROM first_deposits fd
JOIN qualified_players qp ON fd.player_id = qp.player_id
GROUP BY month
ORDER BY month;
"""

#la libreria se encarga de realizar la conexion con la DB y hace la lectura de la consulta
cpa_df = pd.read_sql_query(query_cpa, conn)
```

```python
#Realizamos la impresion de los datos de las consultas
print("FTD por mes:")
print(ftd_df)
print("")

print("CPA por mes:")
print(cpa_df)
print("")
```

```python
#Realizamos la verificacion de la calidad de la informacion de FTD
print("*** Verificación de FTD:*** ")
print("")

#Mostramos resumen tecnico de la consulta
print("===> Estado general de FTD:")
print(ftd_df.info())
print("")

#Mostramos estadisticas que describen los campos de numeros
print("\n===> Estadísticas FTD:")
print(ftd_df.describe())
print("")

#Mostramos los valores nulos existentes por columna
print("\n===> Valores nulos FTD:")
print(ftd_df.isnull().sum())
print("")

#Mostramos los valores duplicados
print("\n===> Duplicados FTD:")
print(ftd_df.duplicated().sum())
```

```python
#Realizamos la verificacion de la calidad de la informacion de CPA
print("*** Verificación de CPA:*** ")
print("")

#Mostramos resumen tecnico de la consulta
print("===> Estado general de CPA:")
print(cpa_df.info())
print("")

#Mostramos estadisticas que describen los campos de numeros
print("\n===> Estadísticas CPA:")
print(cpa_df.describe())
print("")

#Mostramos los valores nulos existentes por columna
print("\n===> Valores nulos CPA:")
print(cpa_df.isnull().sum())
print("")


#Mostramos los valores duplicados
print("\n===> Duplicados CPA:")
print(cpa_df.duplicated().sum())
print("")
```

```python
#Cargamos el archivo de internal records suministrado para validar la informacion
internal_df = pd.read_csv("data/internal_records.csv")
print(internal_df.head())
```

```python
#Realizamos la union de los datos del CSV y del dataframe del query de FTD mensual realizado
merged_ftd = pd.merge(internal_df, ftd_df, on="month", how="left")
print(merged_ftd)
```

```python
#Realizamos la union de los datos del CSV y del dataframe del query de CPA mensual realizado
merged_cpa = pd.merge(internal_df, cpa_df, on="month", how="left")
print(merged_cpa)
```

```python
#Discrepancia detectada de los datos de FDT mensuales de la consulta con el CSV 
merged_ftd["ftd_diff"] = merged_ftd["expected_ftd"] - merged_ftd["ftd_count"]
print(merged_ftd["ftd_diff"])
```

```python
#Discrepancia detectada de los datos de CPA mensuales de la consulta con el CSV 
merged_cpa["cpa_diff"] = merged_cpa["expected_cpa"] - merged_cpa["cpa_count"]
print(merged_cpa["cpa_diff"])
```

```python
# Filtrar meses con diferencia en FTD
discrepancias_ftd = merged_ftd[merged_ftd["ftd_diff"] != 0]
print("Meses con discrepancia en FTD:")
print(discrepancias_ftd)
```

```python
# Filtrar meses con diferencia en CPA
discrepancias_cpa = merged_cpa[merged_cpa["cpa_diff"] != 0]
print("Meses con discrepancia en CPA:")
print(discrepancias_cpa)
```

```python
# Se realiza el grafico de la discrepancia FTD por mes
plt.figure(figsize=(10, 5))
plt.bar(merged_ftd["month"], merged_ftd["ftd_diff"], color="tomato")
plt.title("Diferencia entre FTD esperado y calculado por mes")
plt.xlabel("Mes")
plt.ylabel("Diferencia (FTD)")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Se realiza el grafico de la discrepancia CPA por mes
plt.figure(figsize=(10, 5))
plt.bar(merged_cpa["month"], merged_cpa["cpa_diff"], color="skyblue")
plt.title("Diferencia entre CPA esperado y calculado por mes")
plt.xlabel("Mes")
plt.ylabel("Diferencia (CPA)")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Traemos la información desde la DB filtrando los depositos realizados
query_total_amount = """
SELECT SUBSTR(deposit_date, 1, 7) AS month, SUM(amount) AS total_amount
FROM deposits
GROUP BY month
ORDER BY month;
"""
#Se crea el dataframe para realizar la lectura del Query
total_deposits_df = pd.read_sql_query(query_total_amount, conn)

# Se realiza la grafica del historico de depositos total
plt.figure(figsize=(10, 5))
plt.plot(total_deposits_df["month"], total_deposits_df["total_amount"], marker='o', color='green')
plt.title("Histórico de dinero depositado por mes")
plt.xlabel("Mes")
plt.ylabel("Total depositado (USD)")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Se realiza el query donde se muestra el jugador que mas a depositado
query_top_player = """
SELECT player_id, SUM(amount) AS total_amount
FROM deposits
GROUP BY player_id
ORDER BY total_amount DESC
LIMIT 1;
"""
top_player_df = pd.read_sql_query(query_top_player, conn)
print("Jugador con mayor cantidad depositada:")
print(top_player_df)
```

```python
# Se realiza el query evidenciando el TOP 5 de los jugadores que mas han depositado
query_top5_players = """
SELECT player_id, SUM(amount) AS total_amount
FROM deposits
GROUP BY player_id
ORDER BY total_amount DESC
LIMIT 5;
"""
top5_df = pd.read_sql_query(query_top5_players, conn)

#Se realiza el grafico del Query anteriormente realizado del top 5 de jugadores con mas depositos
plt.figure(figsize=(8, 5))
plt.bar(top5_df["player_id"], top5_df["total_amount"], color='orange')
plt.title("Top 5 jugadores con mayor monto depositado")
plt.xlabel("Player ID")
plt.ylabel("Total depositado (USD)")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#Se realiza la grafica del comparativo esperado versus el calculado en cuanto a los FDT
plt.figure(figsize=(10, 5))
plt.plot(merged_ftd["month"], merged_ftd["expected_ftd"], label="Esperado", marker='o', color='gray')
plt.plot(merged_ftd["month"], merged_ftd["ftd_count"], label="Calculado", marker='x', color='red')
plt.title("Comparación FTD: Esperado vs Calculado")
plt.xlabel("Mes")
plt.ylabel("Cantidad de FTD")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Se realiza la grafica del comparativo esperado versus el calculado en cuanto a los CPA
plt.figure(figsize=(10, 5))
plt.plot(merged_cpa["month"], merged_cpa["expected_cpa"], label="Esperado", marker='o', color='gray')
plt.plot(merged_cpa["month"], merged_cpa["cpa_count"], label="Calculado", marker='x', color='red')
plt.title("Comparación CPA: Esperado vs Calculado")
plt.xlabel("Mes")
plt.ylabel("Cantidad de CPA")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Se realiza el dataframe del trafico
traffic_df = pd.read_csv("data/traffic_sources.csv")
print(traffic_df.head())
```

```python
# Agrupar por trafficker: contar jugadores y sumar el gasto
trafficker_stats = traffic_df.groupby("trafficker").agg({
    "player_id": "count",
    "cost_of_acquisition": "sum"
}).reset_index()

# Renombrar columnas para mayor claridad
trafficker_stats.rename(columns={
    "player_id": "jugadores_generados",
    "cost_of_acquisition": "gasto_total"
}, inplace=True)

# Ordenar por más jugadores
trafficker_stats = trafficker_stats.sort_values(by="jugadores_generados", ascending=False)

# Mostrar resultado
print(trafficker_stats)
```

```python
#Se realiza la exportacion como CSV para poder manejarlo facilmente en BI
trafficker_stats.to_csv("output/estadisticas_trafficker.csv", index=False)
```
